```python
from typing import Optional

# Step 1: Define ListNode class
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# Step 2: Define merge function inside Solution class
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode()
        print("dummy", dummy)
        tail = dummy
        print("tail", tail)

        while list1 and list2:
            if list1.val < list2.val:
                print("list1.val", list1.val)
                print("list2.val", list2.val)
                tail.next = list1
                print("tail.next", tail.next)
                print("dummy", dummy)
                list1 = list1.next
            else:
                print(list1.val, "list1.val")
                print(list2.val, "list2.val")
                tail.next = list2
                print("tail.next", tail.next)
                print(dummy, "dummy")
                list2 = list2.next
            tail = tail.next
            print("tail", tail)

        if list1:
            tail.next = list1
        elif list2:
            tail.next = list2

        print("final tail", tail)
        print("dummy final", dummy)
        return dummy.next

# Step 3: Helper function to build linked list from Python list
def build_linked_list(values):
    if not values:
        return None
    head = ListNode(values[0])
    current = head
    for val in values[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

# Step 4: Helper function to print linked list
def print_linked_list(head):
    while head:
        print(head.val, end=" → ")
        head = head.next
    print("None")

# Step 5: Test the function
list1 = build_linked_list([1, 2, 4])
list2 = build_linked_list([1, 3, 4])

solution = Solution()
merged_head = solution.mergeTwoLists(list1, list2)
print("Merged list:")
print_linked_list(merged_head)
```
